# Numerical analysis and visualisation

```python
df2=df1.groupby(df["organization"]).size()
df2
```

```python
df2.plot.bar(title='pay records per organization',xlabel='organizations',ylabel='pay record count',figsize=(15,7))
```

With the above operation on the given data we got the count of the payroll record per organization

```python
# converting hire_date column to datetime data type

df1['hire_date']=pd.to_datetime(df1['hire_date'],errors='coerce')

# creating hire_year column and assiging year part of date from hire_date column
df1['hire_year']=df1['hire_date'].dt.year
df1

```

```python
# calculating count of employees hired per year for all the organizations  
df5=df1.groupby(['hire_year']).size()

df5
```

```python
df5.plot.bar(title='hiring done per year',xlabel='hiring year',ylabel='count of hirings done',figsize=(15,7))
```

```python
# creating a dataframe with only two columns

df8=df.drop(["first_name","middle_initial","last_name","system","agency","subtitle","class_code","pay_rate","regular_earnings","overtime_earnings","other_earnings","ytd_gross_earnings","hire_date","term_date","suffix "], axis=1)

df8
```

```python
# aggregating the annual salaries per organization

df9=df8.groupby(df["organization"]).sum()

df9
```

```python
df9.plot.bar(title='expenditure of organizations towards salary',xlabel='organization',ylabel='sum of annual salaries',figsize=(15,7))
```

With the above operation on the data we analysed the money spent by each organization on paying salaries annually

```python
df10=df9.squeeze() # converting dataframe to series

df11=pd.concat([df10, df2], axis=1) # concatinating two dataframes

df11

# references: https://www.geeksforgeeks.org/how-to-combine-two-dataframe-in-python-pandas/
# https://datatofish.com/pandas-dataframe-to-series/
```
