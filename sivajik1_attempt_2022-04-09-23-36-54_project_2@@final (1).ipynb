```python
#importing all the required packages
import pandas as pd                   
import matplotlib.pyplot as plt       
import numpy as np                    
import seaborn as sns                 
import datetime         
import plotly.graph_objects as go
```

```python
web= 'https://data.cdc.gov/api/views/9mfq-cb36/rows.csv?accessType=DOWNLOAD' # To fetch data from web
df= pd.read_csv(web) #To read CSV file
df.head(5) #This function displays the first ten rows of the dataframe
```

```python
df.shape  
```

```python
print(df.isnull().sum())                                                       #the total number of null values in each column
```

### Data cleaning

```python
date_frame_1=df.drop(columns=['consent_cases','consent_deaths'])  #columns are being removed
```

```python
date_frame_1.shape     #getting the shape of dataframe
```

```python
data_frame_2=date_frame_1.fillna(date_frame_1.mean(numeric_only=True))                   #replacing null with mean value in that column
```

```python
data_frame_2.isnull().sum()    
```

### Hypotheses:

### 1. The number of cases in texas are high when compare to all states
###      False

```python
#graph of state vs total cases in each state.
plt.figure(figsize=(20,5))
s = sns.barplot(x = "state", y = "tot_cases", data=data_frame_2)
plt.xticks(fontsize='x-small')
```

### False (Caflonia(CA) has the highest cases when compared to all the states not (TX))

### 2. The number of deaths cases in texas are high when compare to all states
###      False

```python
#graph of state vs total deaths in each state.
plt.figure(figsize=(20,5))
S = sns.barplot(x = "state", y = "tot_death", data=data_frame_2)
plt.xticks(fontsize='x-small')
```

### False (Caflonia(CA) has the highest number of death cases not Texas(TX))

### 3. Total number of death cases are higher during the month of may
###  False

```python
#graph of month vs total number of death cases
data_frame_2['submission_date'] = pd.to_datetime(data_frame_2['submission_date'],format='%m/%d/%Y') #fetching month from submission date
data_frame_2['month_from_submission'] = pd.DatetimeIndex(data_frame_2['submission_date']).month
data_frame_2.groupby(["state","month_from_submission"])
sns.lineplot(data=data_frame_2, x="month_from_submission", y="tot_death")
```

###  False(December month has the highest number of death cases not May)

### 4. Total number of cases are higher during the month of may
###  False

```python
#graph of month vs total number of cases
data_frame_2['submission_date'] = pd.to_datetime(data_frame_2['submission_date'],format='%m/%d/%Y') #fetching month from submission date
data_frame_2['month_from_submission'] = pd.DatetimeIndex(data_frame_2['submission_date']).month
data_frame_2.groupby(["state","month_from_submission"])
sns.lineplot(data=data_frame_2, x="month_from_submission", y="tot_cases")
```

###      False(January month has the highest number of cases not may)

###  5. Florida has the highest number of new arising cases
###  False

```python
#graph of state vs new cases in each state.
plt.figure(figsize=(20,5))
s = sns.barplot(x = "state", y = "new_case", data=data_frame_2)
plt.xticks(fontsize='x-small')
```

###      False(CA has the highest number of new arriving cases not florida)

### 6. Total deaths have not increased by more than 80%
### True

```python
#In piechart visualisation, We compare number of cases versus number of deaths
sum_column= data_frame_2.sum(numeric_only=True)
fig = go.Figure(data=[go.Pie(labels=['tot_cases', 'tot_death'], values= sum_column)])
fig.show()
```

###       True(50.8% is total deaths less than 80% )
