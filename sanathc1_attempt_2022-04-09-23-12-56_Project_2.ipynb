<h3>Hypothesis 1</h3>
<h3>The Crime rate in Maryland are more than all other States</h3>

In this Hypothesis we are checking the highest crime rate as per state.
so group by state which gives all the data summation as per state

```python
crime=df.groupby('State').sum()
crime
```

```python
All_rate=crime['Rate.Property.All']+crime['R.Violent.All']     #adding Rate total crime(Property+violent)
```

```python
plt.figure(figsize=(15,4))                                     #Plotting States versus All_rate(Crime)
plt.bar(crime.index,All_rate,color='green')
plt.xticks(rotation=90)
```

```python
All_total=crime['T.Property.All']+crime['T.Violent']          #adding total crime(Property+violent)
```

```python
plt.figure(figsize=(15,4))                         #Plotting States versus All_total(Crime)
plt.bar(crime.index,All_total,color='red')
plt.xticks(rotation=90)
```

<h4>No, We can conclude by above visuals the crime(Rate_crime and property_crime) rate in maryland is not more than other states(District of columbia and united states))</h4>

<h3> Hypothesis-2</h3>
<h3>The Average of total Property Crime is more than the average of violence crime</h3>

```python
Property_rate=int(crime[["Rate.Property.All"]].mean())                       #Calculating the average of rate_property
print(Property_rate)
```

```python
Violent_rate=int(crime[["R.Violent.All"]].mean())                            #Calculating the average of rate_violent
print(Violent_rate)
```

```python
df2=pd.DataFrame([{'Property':Property_rate,'Violent':Violent_rate}])       #storing the average values in new dataframe
df2
```

```python
df2.plot.bar()                                                    #plotting graph of rate property versus  rate violence
```

```python
Property_total=int(crime[["T.Property.All"]].mean())                          #Calculating the average of total_property
print(Property_total)
```

```python
Violent_total=int(crime[["T.Violent"]].mean())                      #Calculating the average of total_violent
print(Violent_total)
```

```python
df3=pd.DataFrame([{'Property':Property_total,'Violent':Violent_total}])      #storing the average values in new dataframe
df3
```

```python
df3.plot.bar()                                    #plotting graph of total property versus total violence
```

<h4>Yes, The average of both rate and total property is grater than the violence</h4>
