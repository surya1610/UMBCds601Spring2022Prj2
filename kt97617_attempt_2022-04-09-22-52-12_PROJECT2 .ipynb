Objective: Classify a new patients who had undergone surgery for breast cancer will survive 5 years or longer OR die within 5 year

```python
counts,bin_edges = np.histogram(data1['Year_of_Operation'], bins=10, density=True)

pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf, label = 'PDF')
plt.plot(bin_edges[1:], cdf, label = 'CDF')
plt.xlabel("Age")
plt.title("CDF plot")
plt.legend()
plt.show()
```

It is observed that around 73% of patients who died within 5 years have age less than or equal to 60

```python
plt.figure(figsize=(6,6))

sns.boxplot(data= data, x='Survival_Status', y='axil_nodes')
plt.title("Box Plot w.r.t Age feature")

blue_patch = mpatch.Patch(color = 'blue', label = 'data1')
orange_patch = mpatch.Patch(color = 'orange', label = 'data2')

plt.legend(handles=[blue_patch, orange_patch])
plt.show()
```
